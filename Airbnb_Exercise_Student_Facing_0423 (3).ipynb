We've included an example of what the steps would look like:

1. Store your results of your .value_counts() operation into a variable 
    
    `variable = dataframe['column'].value_counts().head(20)`
    
    This stores the results of the top 20 results into a variable called condition
    
    <p>
    
2. Access the host_ids that make up the top 20 using the .reset_index() method
    
    `id_list = variable.reset_index()['index'] `
    
    The Value Counts is a `GROUP BY COUNT` implementation. As such, we can reset the index and access the list of host IDs
    that are now stored under the variable called index.
    
    <p>
    
3. Subset your dataframe using the `id_list` that now contains the top 20 host IDs you have extracted

<h3 style="color:teal"> Write your answer below </h3>

````python

top_20_hosts = airbnb_dataframe['host_id'].value_counts().head(20)


id_list = top_20_hosts.index

host_dataframe = airbnb_dataframe[airbnb_dataframe['host_id'].isin(id_list)]
````

<h2 style='color:salmon'> Sample Output </h3>

Now that we've got out dataframe, it's time for us to validate the number of properties each host_id may have.
We're going to validate this using the combination of `host_id` and `name`, assuming that the combination of these two fields, 
will help us understand how many properties each airbnb host might have!

<b> C. Using the `host_dataframe` you created, group the dataframe by `host_id` and `name` to identify the total number of properties each `host_id` may really have using the `.count()` aggregate method. </b>

To successfully answer this question will require a few steps:

1. You'll need to GROUP BY the `host_id` and `name` to get groupings like the below:

````python
host_dataframe.groupby(['host_id','name']).count().head()
````

2. Reset the index after your GROUP BY so `host_id` and `name` are columns once more

    `dataframe.groupby(['host_id','name']).reset_index()`
    <p>
    
3. Lastly, <b> re-group </b> by `host_id`, apply the .count() aggregation and store this in a variable called `host_dataframe_name`!
    <p>
        
The reason we do this is we do this is demonstrated below:

<h3 style="color:teal"> Write your answer below </h3>

````python
# Create a temporary dataframe that identifies unique host/listing name pairs
host_dataframe_unique = host_dataframe.groupby(['host_id', 'name']).size().reset_index()

# Group by host_id and count the occurrences of the 'name' column
host_dataframe_name = host_dataframe_unique.groupby('host_id')['name'].count().reset_index()

# View the result
host_dataframe_name.head()
````

<h2 style='color:salmon'> Sample Output </h3>

Great job so far! We can see that we have some different results from our first `.value_counts()` approach.
Now it's time to visualize our results using a bar chart! 

<b> D. Create a Bar Chart that shows the number of properties each `host_id` has utilizing the `host_dataframe_name` you created in the previous question. </b>

<h3 style="color:teal"> Write your answer below </h3>

````python
import matplotlib.pyplot as plt

# 1. Sort the dataframe so the chart displays the hosts with the most properties at the top
host_dataframe_name = host_dataframe_name.sort_values(by='name', ascending=True)

# 2. Use plt.subplots() to initialize your plot as required by the criteria
fig, ax = plt.subplots(figsize=(12, 8))

# 3. Create a horizontal bar chart (barh)
# We convert host_id to a string so matplotlib treats it as a category label
ax.barh(host_dataframe_name['host_id'].astype(str), host_dataframe_name['name'], color='skyblue')

# 4. Add titles and labels for clarity
ax.set_title('Number of Unique Property Names per Host (Top 20)')
ax.set_xlabel('Count of Unique Property Names')
ax.set_ylabel('Host ID')

# 5. Add gridlines for readability (optional but recommended)
ax.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
````

Whilst we know that a small number of hosts make up the bulk of airbnb properties; could it be that this is spread out uniformly in the different boroughs (i.e. `neighbourhood_group`)?

Let's take a look at understand just how different the number of properties, pricing, and the standard deviation is for the `neighbourhood_group` column in the `airbnb_dataframe` is!

<b> E. Using the .agg() method you learned about previously, calculate the `count`, `mean` and `std` for the `price` column for each `neighbourhood_group`. Upon completion of this, create three bar charts showing the `count`, `mean` and `std` for each `neighbourhood group`.  </b>

Hint: Think about the PSP FS Framework you've learned about to help you create this visualization!

<h3 style="color:teal"> Write your answer below </h3>

````python
import matplotlib.pyplot as plt

# 1. Group by 'neighbourhood_group' and aggregate the 'price' column
# We calculate count, mean, and standard deviation (std)
borough_stats = airbnb_dataframe.groupby('neighbourhood_group')['price'].agg(['count', 'mean', 'std'])

# 2. Create a figure with 3 subplots arranged horizontally (1 row, 3 columns)
# figsize ensures the charts have enough space to be readable
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# 3. Create the first chart: Count of Listings
ax[0].barh(borough_stats.index, borough_stats['count'], color='teal')
ax[0].set_title('Number of Listings')
ax[0].set_xlabel('Count')

# 4. Create the second chart: Average Price
ax[1].barh(borough_stats.index, borough_stats['mean'], color='coral')
ax[1].set_title('Average Price')
ax[1].set_xlabel('Price ($)')

# 5. Create the third chart: Price Standard Deviation
ax[2].barh(borough_stats.index, borough_stats['std'], color='gold')
ax[2].set_title('Price Standard Deviation')
ax[2].set_xlabel('Standard Deviation')

# 6. Apply tight_layout to ensure labels and titles don't overlap
plt.tight_layout()

# Display the charts
plt.show()
````

## Appendix: Python Cheat Sheet 

**SELECTING COLUMNS**

```Python
- dataframe['a'] -- Select column a from a dataframe individually
- dataframe[['a','b']] -- Select columns a and b from a dataframe
- dataframe[:5] -- Select the first five rows from the dataframe
- dataframe[2:] -- Select all the rows of the dataframe starting from row 2 onwards
- dataframe.head(n) -- Select the number of rows in the dataframe specified by n 
- dataframe.loc[row,column] -- Selects the number of rows, and columns specified with labels
- dataframe.iloc[row,column_index] -- Select the number of rows, and columns referencing their index

```

**LIBRARIES**
```Python
- import pandas as pd -- This imports the pandas library that contains the methods you need to manipulate tabular data
- import numpy as np -- This imports the numpy library that contains the np.where (IF) method you need to utilize
- import warnings -- This is the warning library that can be imported to either highlight or hide warning messages
- warnings. filterwarnings('ignore')--This ignores Python warning errors
- import matplotlib.pyplot as plt -- This is the visualization library needed to create plots

```
**OPERATORS**
```Python
- < - Less than
- > - Greater than
- <= - Less than or equal
- >= - Greater than or equal
- <> - Not equal
- = - Equal
- & - AND (e.g. dataframe['Column']>=1) & (dataframe['Column']<=2)
- | - OR
```

**DESCRIPTIVE STATISTICS FUNCTIONS**
```Python
- dataframe['column'].mean() - Returns the average value of a column
- dataframe['column'].min() - Returns the minimum value of a column
- dataframe['column'].max() - Returns the minimum value of a column
- dataframe['column'].sum() - Returns the sum of a column
- dataframe['column'].median() - Returns the median of a column

```

**SUBSETTING COLUMNS**
```Python
- dataframe['column'].isin(['A','B']) - Returns a T/F boolean list of entries matching A and B which are textual fields
- dataframe[condition] - Provided condition is a T/F boolean expression, it will return all matching entries that are
True
```

**USING NP.WHERE() AS A PROXY FOR IF STATEMENTS**
```Python
- np.where(dataframe['column']> 1, "Return when True", "Return when False") -- A True/False expression comparison for numerical terms.
- np.where(dataframe['column'].isin(['A','B']),"Return when True", "Return when False") -- A True/False expression comparison for textual comparisons.
```

**SORTING VALUES**
```Python
- dataframe.groupby(['A']).sum()[['B','C']].sort_values(by=['B'], ascending=False) -- The syntax to sort a dataframe using specific columns
- dataframe.groupby(['A']).sum()['B'].sort_values(ascending=False) -- The syntax to sort a dataframe provided you only have one numerical column to sort on. 

```

**DICTIONARIES AND AGGREGATIONS**
```Python
- Dictionaries are special data types in Python defined by a key-value relationship. 
Syntax wise, they use the {'key': 'value'} syntax
- .agg(['X','Y']) represents a advanced method of providing multiple descriptive statistics where X and Y represents aggregation functions (i.e. sum, mean, min, max etc.) 

```
